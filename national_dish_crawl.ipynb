```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get(url="https://en.wikipedia.org/wiki/National_dish",)
soup = BeautifulSoup(response.content, 'html.parser')

all_spans = soup.find(id="bodyContent").find_all('span')
all_spans.find(id='A')
```
